# Creating DataFrames

```python
import pandas as pd
```

```python
# Specify values for each column
df = pd.DataFrame(
    {"a": [4, 5, 6],
     "b": [7, 8, 9],
     "c": [10, 11, 12]},
    index=[1, 2, 3]
)
df
```

1.png

```python
#Specify values for each row.
# Remember tables is like matrix thats why we use 2D array
df = pd.DataFrame(
    [[4, 7, 10],
     [5, 8, 11],
     [6, 9, 12]],
     index=[1, 2, 3],
     columns=["a", "b", "c"]
)
df
```

1.png

```python
# Create DataFrame with a MultiIndex
df = pd.DataFrame(
    {"a": [4, 5, 6],
     "b": [7, 8, 9],
     "c": [10, 11, 12]
    },
    index=pd.MultiIndex.from_tuples(
        [('d',1), ("d",2),
         ('e',2)], names=["n", "v"]
    )
)
df
```

## Sort

## Rank

```python
"""
df.rank() is a method in Pandas DataFrame that ranks
the values in each column based on their position within the column.
"""
df.rank()
```

```python
# Explaining more about rank function
m = pd.DataFrame({'A': [1, 3, 2], 'B': [4, 2, 1]})
m['A_rank'] = m['A'].rank(method='min')
print(m)
```

#Applying Functions

Explaining lambda function

```python
numbers = [1, 2, 3, 4, 5]

# Double each number in the list
doubled_numbers = list(map(lambda x: x * 2, numbers))
# maping the lambda function operations on a numbers list

print(doubled_numbers)
```

```python
f = lambda x: x*2
df.apply(f) # Apply function
df.applymap(f) #Apply function element wise
```
